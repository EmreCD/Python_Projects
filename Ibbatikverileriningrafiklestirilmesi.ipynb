```python
#Proje: İBB'ye ait atıkların geri dönüşümü verisetinin düzenlendikten sonra kullanılarak grafiklere aktarılması.
#Kod-1 : Çizgi Grafiklerinin ayrı ayrı oluşturulması
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import drive
#google colab drive bağlantısı
drive.mount('/content/drive')
# CSV dosyasını oku
dosya_yolu = '/content/drive/My Drive/VERI_YAPILARI_ISTKA/veriseti/atiktangerikazanimmiktarlari.csv'
#Dosya yolunuzu ve dosyanızın csv olduğundan emin olunuz
csv_dosya = pd.read_csv(dosya_yolu)
yil=csv_dosya.iloc[:,1]
UKM=csv_dosya.iloc[:,2]
GDMM=csv_dosya.iloc[:,3]
ATYM=csv_dosya.iloc[:,4]
CGUEEM=csv_dosya.iloc[:,5]
plt.figure(figsize=(20,15))
plt.subplot(2,2,1) #kompost miktarının
plt.plot(yil, UKM, marker='o', color='blue')
plt.title('Uretilen Kompost Miktarı [ton]')
plt.xlabel('Yıl')
plt.ylabel(csv_dosya.columns[2])
plt.xticks(yil)
plt.grid(True)

plt.subplot(2, 2, 2)
plt.plot(yil, GDMM, marker='o', color='green')
plt.title('IBB Tesislerinde Geri Dönüştürülen Malzeme Miktarı [ton]')
plt.xlabel('Yıl')
plt.ylabel(csv_dosya.columns[3])
plt.xticks(yil)
plt.grid(True)

plt.subplot(2, 2, 3)
plt.plot(yil, ATYM, marker='o', color='red')
plt.title('Atiktan Turetilmis Yakit Miktarı [ton]')
plt.xlabel('Yıl')
plt.ylabel(csv_dosya.columns[4])
plt.xticks(yil)
plt.grid(True)

plt.subplot(2, 2, 4)
plt.plot(yil, CGUEEM, marker='o', color='purple')
plt.title('Cop Gazindan Uretilen Elektrik Enerjisi Miktarları [MWh]')
plt.xlabel('Yıl')
plt.ylabel(csv_dosya.columns[5])
plt.xticks(yil)
plt.grid(True)

# Grafikleri yerleştir ve göster
plt.tight_layout()
plt.show()
```

```python
#Kod-2 : Tek çizgi grafiğinde tüm değerlerin görüntülenmesi
plt.figure(figsize=(20, 18))

plt.plot(yil, UKM, marker='o', label=csv_dosya.columns[2], color='blue')
plt.plot(yil, GDMM, marker='o', label=csv_dosya.columns[3], color='green')
plt.plot(yil, ATYM, marker='o', label=csv_dosya.columns[4], color='red')
plt.plot(yil, CGUEEM, marker='o', label=csv_dosya.columns[5], color='purple')

# Grafiği özelleştir
plt.title('IBB nin Yıllara Göre Atıktan Elde Edilen Verilerin Görselleştirilmesi')
plt.xlabel('Yıl Bilgisi')
plt.ylabel('Üretim Miktarı')
plt.xticks(yil)
#plt.legend()
plt.legend(["Uretilen Kompost Miktarı [ton]", "IBB Tesislerinde Geri Dönüştürülen Malzeme Miktarı [ton]", "Atiktan Turetilmis Yakit Miktarı [ton]", "Cop Gazindan Uretilen Elektrik Enerjisi Miktarları [MWh]"])
plt.grid(True)

# Grafiği göster
plt.show()
```

```python
#Kod-3 : Çubuk grafiğinde tüm değerlerin görüntülenmesi
bar_width = 0.2
index = range(len(yil))

plt.figure(figsize=(16, 14))

# Kompost Miktarı Bar Plot
plt.bar(index, UKM, bar_width, label=csv_dosya.columns[1], color='blue')

# Geri Dönüştürülen Malzeme Miktarı Bar Plot
plt.bar([i + bar_width for i in index], GDMM, bar_width, label=csv_dosya.columns[2], color='green')

# Atıktan Türetilmiş Yakıt Miktarı Bar Plot
plt.bar([i + 2 * bar_width for i in index], ATYM, bar_width, label=csv_dosya.columns[3], color='red')

# Çöp Gazından Üretilen Elektrik Enerjisi Miktarı Bar Plot
plt.bar([i + 3 * bar_width for i in index], CGUEEM, bar_width, label=csv_dosya.columns[4], color='purple')

# X eksenini belirli yıllara ayarla
plt.xlabel('Yıl')
plt.ylabel('Miktar')
plt.title('Atıktan Geri Kazanım Miktarları')
plt.xticks([i + 1.5 * bar_width for i in index], yil)
plt.legend(["Uretilen Kompost Miktarı [ton]", "IBB Tesislerinde Geri Dönüştürülen Malzeme Miktarı [ton]", "Atiktan Turetilmis Yakit Miktarı [ton]", "Cop Gazindan Uretilen Elektrik Enerjisi Miktarları [MWh]"])
plt.grid(True)

plt.tight_layout()
plt.show()
```
