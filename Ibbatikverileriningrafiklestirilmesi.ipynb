```python
#Proje: BB'ye ait atklarn geri dnm verisetinin dzenlendikten sonra kullanlarak grafiklere aktarlmas.
#Kod-1 : izgi Grafiklerinin ayr ayr oluturulmas
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import drive
#google colab drive balants
drive.mount('/content/drive')
# CSV dosyasn oku
dosya_yolu = '/content/drive/My Drive/VERI_YAPILARI_ISTKA/veriseti/atiktangerikazanimmiktarlari.csv'
#Dosya yolunuzu ve dosyanzn csv olduundan emin olunuz
csv_dosya = pd.read_csv(dosya_yolu)
yil=csv_dosya.iloc[:,1]
UKM=csv_dosya.iloc[:,2]
GDMM=csv_dosya.iloc[:,3]
ATYM=csv_dosya.iloc[:,4]
CGUEEM=csv_dosya.iloc[:,5]
plt.figure(figsize=(20,15))
plt.subplot(2,2,1) #kompost miktarnn
plt.plot(yil, UKM, marker='o', color='blue')
plt.title('Uretilen Kompost Miktar [ton]')
plt.xlabel('Yl')
plt.ylabel(csv_dosya.columns[2])
plt.xticks(yil)
plt.grid(True)

plt.subplot(2, 2, 2)
plt.plot(yil, GDMM, marker='o', color='green')
plt.title('IBB Tesislerinde Geri Dntrlen Malzeme Miktar [ton]')
plt.xlabel('Yl')
plt.ylabel(csv_dosya.columns[3])
plt.xticks(yil)
plt.grid(True)

plt.subplot(2, 2, 3)
plt.plot(yil, ATYM, marker='o', color='red')
plt.title('Atiktan Turetilmis Yakit Miktar [ton]')
plt.xlabel('Yl')
plt.ylabel(csv_dosya.columns[4])
plt.xticks(yil)
plt.grid(True)

plt.subplot(2, 2, 4)
plt.plot(yil, CGUEEM, marker='o', color='purple')
plt.title('Cop Gazindan Uretilen Elektrik Enerjisi Miktarlar [MWh]')
plt.xlabel('Yl')
plt.ylabel(csv_dosya.columns[5])
plt.xticks(yil)
plt.grid(True)

# Grafikleri yerletir ve gster
plt.tight_layout()
plt.show()
```

```python
#Kod-2 : Tek izgi grafiinde tm deerlerin grntlenmesi
plt.figure(figsize=(20, 18))

plt.plot(yil, UKM, marker='o', label=csv_dosya.columns[2], color='blue')
plt.plot(yil, GDMM, marker='o', label=csv_dosya.columns[3], color='green')
plt.plot(yil, ATYM, marker='o', label=csv_dosya.columns[4], color='red')
plt.plot(yil, CGUEEM, marker='o', label=csv_dosya.columns[5], color='purple')

# Grafii zelletir
plt.title('IBB nin Yllara Gre Atktan Elde Edilen Verilerin Grselletirilmesi')
plt.xlabel('Yl Bilgisi')
plt.ylabel('retim Miktar')
plt.xticks(yil)
#plt.legend()
plt.legend(["Uretilen Kompost Miktar [ton]", "IBB Tesislerinde Geri Dntrlen Malzeme Miktar [ton]", "Atiktan Turetilmis Yakit Miktar [ton]", "Cop Gazindan Uretilen Elektrik Enerjisi Miktarlar [MWh]"])
plt.grid(True)

# Grafii gster
plt.show()
```

```python
#Kod-3 : ubuk grafiinde tm deerlerin grntlenmesi
bar_width = 0.2
index = range(len(yil))

plt.figure(figsize=(16, 14))

# Kompost Miktar Bar Plot
plt.bar(index, UKM, bar_width, label=csv_dosya.columns[1], color='blue')

# Geri Dntrlen Malzeme Miktar Bar Plot
plt.bar([i + bar_width for i in index], GDMM, bar_width, label=csv_dosya.columns[2], color='green')

# Atktan Tretilmi Yakt Miktar Bar Plot
plt.bar([i + 2 * bar_width for i in index], ATYM, bar_width, label=csv_dosya.columns[3], color='red')

# p Gazndan retilen Elektrik Enerjisi Miktar Bar Plot
plt.bar([i + 3 * bar_width for i in index], CGUEEM, bar_width, label=csv_dosya.columns[4], color='purple')

# X eksenini belirli yllara ayarla
plt.xlabel('Yl')
plt.ylabel('Miktar')
plt.title('Atktan Geri Kazanm Miktarlar')
plt.xticks([i + 1.5 * bar_width for i in index], yil)
plt.legend(["Uretilen Kompost Miktar [ton]", "IBB Tesislerinde Geri Dntrlen Malzeme Miktar [ton]", "Atiktan Turetilmis Yakit Miktar [ton]", "Cop Gazindan Uretilen Elektrik Enerjisi Miktarlar [MWh]"])
plt.grid(True)

plt.tight_layout()
plt.show()
```
